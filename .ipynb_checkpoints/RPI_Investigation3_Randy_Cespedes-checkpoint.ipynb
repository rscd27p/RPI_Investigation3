### Types of Synthetic Data

There are two main types of synthetic data:

-**Fully Synthetic:** This dataset does not contain any orginal data (acquired, generated by events, etc.).<br/>
-**Partially Synthetic:** Based on [4] in this type of dataset is the one in which onluy sensitive data is replaced with synthetic data. For example, data that is protected by privacy laws.<br/>


### Techniques for Synthetic Data Generation

Based on [1][4][6], there are three main techniques for synthethic data generation:

* **Generating data according to a distribution**: This case is commong when using **fully synthetic** generated data. Since there is not real data its necessary to get an expert criteria on how the data would look like. For example, it is important to understand the type of distribution for the data. 

Several common types of distribution are described in [5]. In the image below there is an example of these types of distributions.

```python
from IPython.display import Image
Image("Images/Different types of distribution.png") 

#Source: [5]
```

* **Age-based modeling:** the data is generated by a model constructed based a model constructed on real data. When using real data is very important to understand if the data is an acurrate representation of the real behaviour of the system. If the data is not captured taking this into consideration, the data could be wronly generated.<br>

* **Using Deep Learning Techniques:** Some of these types of techniques are described in [6]. For example:<br><br>

  * **Variational Autoencoder (VAE):** This is an unsupervised learning method where an **autoencoder** is used to generate new data by introducing random information in what is known as _latent space_. The autoencoder has two main components  an **encoder**, which compresses real data, and then a **decoder** needs to reconstruct the data as close as possible to the original data [9].
