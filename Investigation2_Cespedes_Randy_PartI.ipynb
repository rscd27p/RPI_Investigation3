Based on the results above, it is clear that there is large amount of outliers in the different airlines. It was decided to use the IRQ technique used in the first investigation [here](https://github.com/rscd27p/Investigacion_1_RP). This technique is named. In this case, the same procedure would be done with the Departure Delay time.

This technique is described [here](https://www.pluralsight.com/guides/cleaning-up-data-from-outliers). The equation used equations are:


$ FD: Flight Data $ <br/>
$ FD_{IRQ}: Interquantile Range $ <br/>
$ \bar{FD}: Flight Data Median $ <br/>
$ FD_{Q1}: Flight Data First Quantile $ <br/>
$ FD_{Q3}: Flight Data Third Quantile $ <br/><br/>

$ FD_{IRQ} =  FD_{Q3} - FD_{Q1}$ <br/><br/>

$ L: Lower Limit $ $ L = \bar{FD} - 1.5 * FD_{IRQ} $ <br/> 
$ U: Upper Limit $ $ L = \bar{FD} + 10 * FD_{IRQ} $ <br/>

**Note:** The upper limit of 10*IRQ was decided after experimentation.

$ FD_{IRQ_range} = [L:U]$

It was decided to estimate how many elements are above of the IRQ Range, to try to identify if too many elements were going to be eliminated.

```python
# To do this process we need to use the third quantile calculated above. Also, the first quantile, the median and
# the interquantile range.

# The interquantile range was multiplied by 10 instead of 1.5 to reduce the amount of data to be dropped.
FD_IRQ = DEP_DELAY_Q3 - simplified_FD_clean['DEP_DELAY'].quantile(.25)
lower_IQR = simplified_FD_clean['DEP_DELAY'].median()-(1.5*FD_IRQ)
upper_IQR = simplified_FD_clean['DEP_DELAY'].median()+(10*FD_IRQ)
FD_IRQ = simplified_FD_clean[simplified_FD_clean['DEP_DELAY'].between(lower_IQR,upper_IQR,inclusive=True)]

print("Total number of flights after processing: " +str(FD_IRQ['DEP_DELAY'].count()))
```

Out of the initial 232 831 flights, we end up with 231 831.

$Total Flights_{deleted} = 232 831 - 231 831 = 1000$

As it can be seen above, each one of these columns have over 360 unique values. Meaning that performing a graphical analysis like the ones above can be hard. One of the ideas is to perform an alysis it is possible to analysis the amount of flights per class. The code below has that objective:

```python
# Define variables to be reusedTotal number of flights after processing: 535078

FD_ORIGIN_counts = FD_IRQ[['ORIGIN']].value_counts()
FD_OCName_counts = FD_IRQ[['ORIGIN_CITY_NAME']].value_counts()
FD_DEST_counts = FD_IRQ[['DEST']].value_counts()
FD_DCName_counts = FD_IRQ[['DEST_CITY_NAME']].value_counts()

# Create Box Plots for Dataset

Num_Flights_per_category = [FD_ORIGIN_counts, FD_OCName_counts, FD_DEST_counts, FD_DCName_counts]
plt.figure(figsize=(20,12))
g = sns.boxplot(data=Num_Flights_per_category, linewidth=2.5).set(xlabel='Category', ylabel='Number of instances per unique subclass')
plt.xticks(ticks = [0,1,2,3], labels = ['ORIGIN','ORIGIN_CITY_NAME','DEST','DEST_CITY_NAME'])
plt.show()
```

In the plot above it is possible to see that there is a large majority of origin and destination airports and cities with a small number flights. **Important:** This does not mean that they represent the majority of the flights. If we check the first and third quantiles of the Origin, we could understand the data better.

```python
O_Q3 = FD_ORIGIN_counts.quantile(0.75)
O_Q1 = FD_ORIGIN_counts.quantile(0.25)
print("Number of instances of ORIGIN Firt quantile: " + str(O_Q1))
print("Number of instances of ORIGIN Third quantile: " + str(O_Q3))
```

This means that 25% of the origin airport names are mentioned less than 64 times each. 64 presents only 0.25% out of 26 169, which the most mentioned airport of origin. The code below creates a list of each category of the flights to be removed.

```python
# Calcualte first Quantile for each category

OCN_Q1 = FD_OCName_counts.quantile(0.25)
D_Q1 = FD_DEST_counts.quantile(0.25)
DCN_Q1 = FD_DCName_counts.quantile(0.25)

# Keep Values on First Quantile to later remove them:

FD_ORIGIN_tbremoved = FD_ORIGIN_counts[FD_ORIGIN_counts < O_Q1] 
FD_OCName_tbremoved = FD_OCName_counts[FD_OCName_counts < OCN_Q1]
FD_DEST_tbremoved = FD_DEST_counts[FD_DEST_counts < D_Q1]
FD_DCName_tbremoved = FD_DCName_counts[FD_DCName_counts < DCN_Q1]
```

As an example in the column below you can observe the result of one of the procedures above. 

```python
print(FD_ORIGIN_tbremoved)
```

All of the Airports mentioned above will be removed. Here is a comparrisson of the data.

Now that there is a list of Aiports and Cities to remove from the data frame. It is necessary to create the code to remove any flight that belong to these. However, the _FD_ORIGIN_tbremoved_ is a Multindex, a data type from pandas, that contains a lot of information. What it is needed for this step is the tuple index. However, its format would make it harder to use it.

The code below creates simple lists with the airport and city names to remove from the data set.

```python
# Adapt lists

FD_ORIGIN_tbremoved_l = []
for idx in range(0,len(FD_ORIGIN_tbremoved.index)):
    FD_ORIGIN_tbremoved_l.append(FD_ORIGIN_tbremoved.index[idx][0])
 
FD_OCName_tbremoved_l = []
for idx in range(0,len(FD_OCName_tbremoved.index)):
    FD_OCName_tbremoved_l.append(FD_OCName_tbremoved.index[idx][0])

FD_DEST_tbremoved_l = []
for idx in range(0,len(FD_DEST_tbremoved.index)):
    FD_DEST_tbremoved_l.append(FD_DEST_tbremoved.index[idx][0])

FD_DCName_tbremoved_l = []
for idx in range(0,len(FD_DCName_tbremoved.index)):
    FD_DCName_tbremoved_l.append(FD_DCName_tbremoved.index[idx][0])



FD_IRQ = FD_IRQ[~FD_IRQ['ORIGIN'].isin(FD_ORIGIN_tbremoved_l)]
FD_IRQ = FD_IRQ[~FD_IRQ['ORIGIN_CITY_NAME'].isin(FD_OCName_tbremoved_l)]
FD_IRQ = FD_IRQ[~FD_IRQ['DEST'].isin(FD_DEST_tbremoved_l)]
FD_IRQ = FD_IRQ[~FD_IRQ['DEST_CITY_NAME'].isin(FD_DCName_tbremoved_l)]

# Count the number of elements after the procedure

print("Total number of flights after processing: " +str(FD_IRQ['DEP_DELAY'].count()))
```
